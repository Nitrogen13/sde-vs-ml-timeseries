```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("percent_results2.csv")
df
```

```python
df = df[df["sigma"] == 0.5][["kappa", "theta_a", "theta_b", "error_kappa", "error_theta_a", "error_theta_b"]]
df
```

```python
param_name_1 = "kappa"
param_name_2 = "theta_b"
target = "theta_a"

unique_values_1 = df[param_name_1].unique()
unique_values_2 = df[param_name_2].unique()

for param_1 in unique_values_1:
    for param_2 in unique_values_2:
        dataframe = df[df[param_name_1] == param_1][df[param_name_2] == param_2]
        x = dataframe[target]
        y = dataframe["error_" + target] 
        plt.figure() 
        plt.title(f"{param_name_1} = {param_1}, {param_name_2} = {param_2}")
        plt.plot(x,y)
        plt.show()
        
```
